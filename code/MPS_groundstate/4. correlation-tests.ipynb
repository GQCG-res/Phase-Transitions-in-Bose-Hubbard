```julia
include("4.a. Correlations.jl")
using Plots
```

```julia
C_MI = correlation_function(0.1, 1., 0., 128)

plot(C_MI; yscale=:log)
```

```julia
C_SF = correlation_function(0.5, 1., 0., 128; sweeps=5)

plot(C_SF; xscale=:log, yscale=:log)
```

First ~10 sites are boundary effects. The plateau is a power law, followed by a long exponential tail from the limited bond dimension.

```julia
density_MI = density_function(0.1, 1., 0., 128)

plot(density_MI)
```

For the Mott insulator, the boundary effects are very small.

```julia
density_SF = density_function(0.5, 1., 0., 128)

plot(density_SF)
```

For the superfluid phase it is a lot more noticable.
