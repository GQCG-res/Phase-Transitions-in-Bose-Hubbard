```python
import prototypes.BoseHubbardModel as bh
import numpy as np
import pandas as pd
import scipy.optimize as opt
import plotly
import plotly.graph_objects as go
import time
import json
```

```python
p = 1.5672020623115244
file = [[float(j) for j in i.strip().split()] for i in open("./MPS_groundstate/correlation functions.txt") if not i.startswith("#")]
l = file[0]

figure = go.Figure()
colors = plotly.colors.DEFAULT_PLOTLY_COLORS

i=0
x = np.linspace(0, len(l), 200)
y = np.exp(-x/p)

figure.add_trace(go.Scatter(
    name = 'Γ(r)',
    mode = 'markers',
    marker=dict(color=colors[i].replace(")", "").replace("rgb", "rgba")+', 1)', size=5),
    x = np.arange(0, len(l)),
    y = l))

figure.add_trace(go.Scatter(
    name = 'exp fit',
    mode = 'lines',
    line=dict(color='rgba(0, 0, 0, 0.5)', width=3),
    x = x,
    y = y))

figure.update_xaxes(title_text='r',
                    ticks="inside",
                    nticks = 5, showgrid=True, zeroline=True, linecolor="black", gridcolor="gray", zerolinecolor="black", tickfont=dict(size=24), tickformat=".0f")
figure.update_yaxes(title_text='correlation',
                    ticks="inside", showgrid=True, zeroline=True, linecolor="black", gridcolor="gray", zerolinecolor="black", tickfont=dict(size=24), exponentformat="power", type="log")
figure.update_layout(paper_bgcolor='rgba(255, 255, 255, 1)', plot_bgcolor='rgba(255, 255, 255, 1)')
figure.update_layout(font=dict(size=26))
figure.update_layout(height = 800, width = 1200)

figure.show()

figure.write_image("./figures/Correlations-MI1.pdf")
time.sleep(5)
figure.write_image("./figures/Correlations-MI1.pdf")
```

```python
a = [64, 128, 256]
file = [[float(j) for j in i.strip().split()] for i in open("./MPS_groundstate/different lengths.txt") if not i.startswith("#")]

figure = go.Figure()
colors = plotly.colors.DEFAULT_PLOTLY_COLORS

for i, l in enumerate(file):
    figure.add_trace(go.Scatter(
        name= 'system length='+str(a[i]),
        mode = 'markers+lines',
        marker=dict(color=colors[i].replace(")", "").replace("rgb", "rgba")+', 1)', size=3),
        line=dict(color=colors[i].replace(")", "").replace("rgb", "rgba")+', 0.3)', width=3),
        x = np.arange(1,len(l)+1),
        y = l)) 

figure.update_xaxes(title_text='r',
                    ticks="inside",
                    nticks = 5, showgrid=True, zeroline=True, linecolor="black", gridcolor="gray", zerolinecolor="black", tickfont=dict(size=24), tickformat=".0f", type="log")
figure.update_yaxes(title_text='correlation',
                    ticks="inside", showgrid=True, zeroline=True, linecolor="black", gridcolor="gray", zerolinecolor="black", tickfont=dict(size=24), tickformat=".2f", type="log")
figure.update_layout(paper_bgcolor='rgba(255, 255, 255, 1)', plot_bgcolor='rgba(255, 255, 255, 1)')
figure.update_layout(font=dict(size=26))
figure.update_layout(height = 800, width = 1200)

figure.show()

figure.write_image("./figures/Correlations-lengths.pdf")
time.sleep(5)
figure.write_image("./figures/Correlations-lengths.pdf")
```

```python
p = 1.5672020623115244
l = [float(i.strip()) for i in open("./MPS_groundstate/fractional density.txt") if not i.startswith("#")]

figure = go.Figure()
colors = plotly.colors.DEFAULT_PLOTLY_COLORS

i=0

figure.add_trace(go.Scatter(
    name = 'Γ(r)',
    mode = 'markers+lines',
    marker=dict(color=colors[i].replace(")", "").replace("rgb", "rgba")+', 1)', size=5),
    line=dict(color=colors[i].replace(")", "").replace("rgb", "rgba")+', 0.3)', width=5),
    x = np.arange(0, len(l)),
    y = l))

x = np.linspace(0, 64, 200)
a = [-1.076826736424263, 0.057787465165711545]
y = 0.75 + a[1] * (x+1)**a[0]

figure.add_trace(go.Scatter(
    name = 'exp fit',
    mode = 'lines',
    line=dict(color='black', width=3),
    x = x,
    y = y))

figure.update_xaxes(title_text='L',
                    ticks="inside",
                    nticks = 5, showgrid=True, zeroline=True, linecolor="black", gridcolor="gray", zerolinecolor="black", tickfont=dict(size=24), tickformat=".0f")
figure.update_yaxes(title_text='ρ (a.u.)',
                    ticks="inside", showgrid=True, zeroline=True, linecolor="black", gridcolor="gray", zerolinecolor="black", tickfont=dict(size=24), tickformat=".2f")
figure.update_layout(paper_bgcolor='rgba(255, 255, 255, 1)', plot_bgcolor='rgba(255, 255, 255, 1)')
figure.update_layout(font=dict(size=26))
figure.update_layout(height = 800, width = 1200)
figure.update_layout(showlegend=False)

figure.show()

figure.write_image("./figures/Density-profiles-fractional density.pdf")
time.sleep(5)
figure.write_image("./figures/Density-profiles-fractional density.pdf")
```
